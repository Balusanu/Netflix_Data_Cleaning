## Netflix Data cleaning

In this Project we are going to load the "netflix1" CSV file & perform following Data cleaning steps
1) Removing redundant columns
2) Renaming columns
3) Dropping duplicates
4) removing NaN( null) values from Data
5) Check for some more transormations

```python
## importing the Pandas library
import pandas as pd
```

```python
## loading the CSV file
data = pd.read_csv("netflix1.csv")
```

```python
data
```

```python
data.shape ## Data has 10 columns & 8790 rows
```

```python
data.head() ##Shows first five rows with columns
```

```python
data.info()
```

1) Removing redundant columns

```python
data.columns ##shows all the column names
```

```python
data.drop(columns="rating",inplace=True) ##we are dropping rating column as it is not much use
```

```python
data
```

2) Renaming columns

```python
data.columns ## we need to capitalise each column name
```

```python
## we can do it using a loop
new_columns_names = []
for i in data.columns:
   new_columns_names.append(i.capitalize())

data.columns = new_columns_names
```

```python
data.columns
```

```python
data.head()
```

3) Dropping duplicates

```python
## checking for duplicate entries in the data
data.duplicated().sum() ## There are no duplicate entries in the data
```

4) removing NaN( null) values from Data

```python
## checking for NaN entries in the data
data.isna().sum() ## There are no NaN entries in the data
```

5) Check for some more transormations

```python
## Removing "s" string from Show_id column
data["Show_id"].unique()
```

```python
data["Show_id"] = data["Show_id"].apply(lambda x: x.split("s")[1])
```

```python
data
```

```python
## changing the  data type of Show_id to int
type(data["Show_id"][0])
```

```python
data["Show_id"] = data["Show_id"].astype(int)
```

```python
type(data["Show_id"][0])
```

```python
data.head()
```

```python
## replacing "/" in Date_added column with "-"
data["Date_added"] = data["Date_added"].apply(lambda x:x.replace("/","-"))
```

```python
## We are going to export the cleaned in CSV & XLSX format
data.to_csv("Netflix_Cleaned_data.csv",index=False)
data.to_excel("Netflix_Cleaned_data.xlsx",index=False)
```
